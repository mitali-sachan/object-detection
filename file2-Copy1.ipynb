```python
import numpy as np
import cv2
import os
import random
import matplotlib.pyplot as plt
```

```python
DIRECTORY= R'C:\Users\hp\OneDrive\Desktop\mitali\project\dataset\training'
CATEGORIES=['forks','glasses','plates','spoons']
```

```python
DIRECTORY_test= R'C:\Users\hp\OneDrive\Desktop\mitali\project\dataset\testing'
CATEGORIES_test=['forks','glasses','plates','spoons']
```

```python
img_size=256
data=[]
for category in CATEGORIES:
    folder=os.path.join(DIRECTORY,category)
    label=CATEGORIES.index(category)
    for img in os.listdir(folder):
        img_path=os.path.join(folder, img)
        img_arr=cv2.imread(img_path)
        img_arr=cv2.resize(img_arr,(img_size,img_size))
        data.append([img_arr,label])
```

```python
img_size=256
datas=[]
for category in CATEGORIES_test:
    folder=os.path.join(DIRECTORY_test,category)
    label=CATEGORIES_test.index(category)
    for img in os.listdir(folder):
        img_path=os.path.join(folder, img)
        img_arr=cv2.imread(img_path)
        img_arr=cv2.resize(img_arr,(img_size,img_size))
        datas.append([img_arr,label])
```

```python
len(data)
```

```python
len(datas)
```

```python
random.shuffle(data)
random.shuffle(datas)
```

```python
x=[]
y=[]
for features,labels in data:
    x.append(features)
    y.append(labels)
```

```python
x_test=[]
y_test=[]
for features,labels in datas:
    x_test.append(features)
    y_test.append(labels)
```

```python
x=np.array(x)
y=np.array(y)
```

```python
x_test=np.array(x_test)
y_test=np.array(y_test)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Flatten,Dense
from keras.callbacks import TensorBoard
from keras.utils import normalize,to_categorical
from keras.preprocessing.image import ImageDataGenerator
from keras.layers import BatchNormalization
import scipy
```

```python
x=normalize(x,axis=1)
x_test=normalize(x_test,axis=1)
y=to_categorical(y)
y_test=to_categorical(y_test)
```

```python
x.shape
x_test.shape
```

```python
train_datagen=ImageDataGenerator(rotation_range=45,width_shift_range=0.2,zoom_range=0.2,horizontal_flip=True)
train_datagen.fit(x)
```

```python
train_generator=train_datagen.flow(x,y,batch_size=16)
```

```python
model=Sequential()

model.add(Conv2D(16,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(128,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(256,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Flatten())
model.add(Dense(128,input_shape=x.shape[1:],activation='relu'))
model.add(Dense(4,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(train_generator,steps_per_epoch=10,epochs=100,validation_data=(x,y))
```

```python
model.save("terms.h5")
```

```python
from PIL import Image
from keras.models import load_model
import cv2
```

```python
model=load_model("terms.h5")
```

```python
fontFace=cv2.FONT_HERSHEY_SIMPLEX
```

```python
video=cv2.VideoCapture(0)
while True:
    _,frame=video.read()
    im=Image.fromarray(frame,'RGB')
    im=im.resize((256,256))
    img_array=np.array(im)
    img_array=img_array/255
    img_array=np.expand_dims(img_array,axis=0)
    prediction=model.predict(img_array)
    prediction=np.argmax(prediction,axis=1)
    PV=np.amax(prediction)
    if prediction==0:
        cv2.putText(frame,str('forks'),(180,75),fontFace,0.75,(0,0,255),5,cv2.LINE_AA)
    elif prediction==1:
        cv2.putText(frame,str('glasses'),(20,35),fontFace,0.75,(0,0,255),5,cv2.LINE_AA)
    elif prediction==2:
        cv2.putText(frame,str('plates'),(20,35),fontFace,0.75,(0,0,255),5,cv2.LINE_AA)
    elif prediction==3:
        cv2.putText(frame,str('spoons'),(20,35),fontFace,0.75,(0,0,255),5,cv2.LINE_AA)
    cv2.imshow("Capturing",frame)
    key=cv2.waitKey(1)
    if key==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
    
```
